```python
import torch.nn as nn
sample = torch.tensor([10.,10.,10.])
# Linear transformation
# This is important because the number of neurons between each layer is different. This allows transformation of tensor into other shapes so they
# can be processed by the next layer.
# For eaxmple, nn.Linear(3,5) takes a tensor of size 3, multiply it by a weight matrix of size (5, 3), producing a tensor of size 5. This allows
# the model to learn relationships between features. The weight matrix is initialized to random values but eventually converges to a fixed value.
linear = nn.Linear(3, 3, bias = False)
print(linear(sample))
```

```python
import torch.nn.functional as F

tensor = torch.tensor([1.,2.,3.])
# Applying the softmax function on the tensor
# The softmax function exponentiates each element x in the tensor to 2.71^x and divides each of them by the exponentiated total sum.
softmax_output = F.softmax(tensor, dim = 0)
print(softmax_output)
```

```python
# Initialize an embedding layer
vocab_size = 1000
embedding_dim = 100
embedding = nn.Embedding(vocab_size, embedding_dim)

# Create some input indices
input_indices = torch.LongTensor([1, 5, 3, 2])

# Apply the embedding layer
embedded_outputs = embedding(input_indices)

# The output will be a tensor of shape (4, 100) where 4 is the number of input indices and 100 is the dimension of the embedding vector
# Embedding is required to transform tokens into a numerical representation that somehow carries their semantics (so into vectors).
print(embedded_outputs.shape)
```

```python
a = torch.Tensor([[1,2], [3,4], [5,6]])
b = torch.Tensor([[7,8,9], [10, 11, 12]])
# dot product between matrices
# The size of the two matrices must be a * b and b * c, and the resulting matrix will have size a * c.
print(a @ b)
print(torch.matmul(a, b))
```

```python
# The following are some common activation functions.
# When forward training the model, use the function from torch.nn not from F.
x = torch.tensor([-0.05], dtype=torch.float32)
# ReLU function returns 0 if the input is < 0, and returns input itself otherwise
y = F.relu(x)
print(y)
# Sigmoid function returns 1/(1+exp(-x))
# Output ranges from 0 to 1
y = F.sigmoid(x)
print(y)
# Tanh function returns (exp(x)-exp(-x))/(exp(x)+exp(-x))
# Output ranges from -1 to 1
y = F.tanh(x)
print(y)

x = torch.tensor([0.05], dtype=torch.float32)
y = F.relu(x)
print(y)
y = F.sigmoid(x)
print(y)
y = F.tanh(x)
print(y)
```
